### <font color=red>Lists</font>
The list is a most versatile datatype available in Python which can be written as a list of comma-separated values (items) between square brackets. Important thing about a list is **that items in a list need not be of the same type.**

**Ex1: Split a string into a list based on a specified separator**

```python
original_string="Hello, how are you today?"

word_list=original_string.split(",")
print("Splitting by space: ",word_list)
```

**Ex2: Making a list of chars in a string**

```python
s="spam"
t=list(s)
print(t)
```

**Ex3: Joining elements of a list into a string**

```python
t=["programming","is","fun"]
#delimiter= " "
print(" ".join(t))
```

**Ex4: Accessing Values in Lists by Index**

```python
list=[1,2,3,4,5,"apple","banana"]

print(list[0])
print(list[-2])
print(list[1:3])
```

**Ex5:Updating Lists**

```python
list=[1,2,3,4,5,6]
print("The value at index 2:", list[2])

list[2]=6
print("The new value at index 2:",list[2])
```

**Ex6: Deleting List Elements**

```python
list=[1,2,3,4,5,6]
print(list)

del list[2]
print(list)
```

### <font color=red>Tuples</font> 
* A tuple is a sequence of **immutable** Python objects. Tuples are sequences, just like lists.<br> 
* What are the differences between tuples and lists ?<br>
* A tuple is a sequence of immutable Python objects. Tuples are sequences, just like lists.
* The differences between tuples and lists are, 
> * the **tuples cannot be changed** unlike lists, 
> * tuples use parentheses, whereas lists use square brackets.
* **Creating a tuple:** tup=(1,2,3,4,5)

**Ex1: Tuples: Accessing Values in Tuples**

```python
tup1 = ('physics', 'chemistry', 1997, 2000) 
tup2 = (1, 2, 3, 4, 5, 6, 7 ) 
print ("tup1[0]: ", tup1[0]) 
print ("tup2[1:5]: ", tup2[1:5])
```

**Updating tuples**


```python
tup1 = (12, 34.56) 
tup2 = ('abc', 'xyz') 
tup3 = tup1 + tup2 
print (tup3)
```

### <font color=red>Dictionaries</font> 
* Dictionary as an unordered set of key: value pairs, with the requirement that the keys are **unique** (within one dictionary).<br>
* Unlike sequences, which are indexed by a range of numbers, dictionaries are indexed by keys, which can be any immutable type.

**Ex1: Creating a dictionary**

```python
dict={"Name":"Zara","Age":7,"Class":"First"}
print(dict)
```

**Ex2:Accessing Values in a Dictionary:**

```python
dict={"Name":"Zara","Age":7,"Class":"First"}

print(dict["Name"])
```

**Ex3: Updating a dictionary**

```python
dict={"Name":"Zara","Age":7,"Class":"First"}
dict["Age"]=8
dict["School"]="DPS School"

print(dict)
```
