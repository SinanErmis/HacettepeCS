#### * Example 11*: **  control flow continued **:
what does this program do?

```python

username = input("Enter your username: ")
password = input("Enter your password: ")

total = len(username) + len(password)

message = "Your username and password has a total of {} characters!"
print(message.format(total))

if total > 40:
    print("The total lenght of username and password should not exceed 40 characters.")
else:
    print("Welcome to the system.")
```

#### * Example 12*: **  control flow continued-checking if two numbers are divisible **

Note: ‘%‘ is **a modulus operator** which divides the left hand operand by the right hand operand and **returns the remainder.** 

```python
number1=int(input("Please enter number to be divided:"))
number2=int(input("Please enter divisor:"))
if(number1%number2==0):
    print("{} can divide {} without a remainder".format(number2,number1))
else:
    print("{} can not divide {} without a remainder".format(number2,number1))
```
